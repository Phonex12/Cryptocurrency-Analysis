### **Exploratory Data Analysis**

**Importing Modules**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set(style = 'darkgrid')
```

**Importing data and Performing Cleanups**

```python
df = pd.read_csv('D:\Crypto\Data\crypto_date.csv')
df.drop(columns='Unnamed: 0',inplace = True)
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df = df.dropna()
df['year'] = df['year'].astype('Int64')
```

```python
df = df[df['name'] == 'Bitcoin'].copy()
```

```python
df_2024 = df[df['year'] == 2024]
```

```python
df_2024 = df_2024.reset_index()
```

```python
df_2024.drop(columns = 'index',inplace=True)
```

```python
df_2024['month_no'] = df_2024['date'].dt.month
```

```python
df_2024['month_name'] = pd.to_datetime(df_2024['month_no'],format = '%m').dt.month_name()
```

```python
df_2024
```

**Plotting the Data**

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df_2024,x = 'month_name',y = 'price',marker = 'o')
plt.xticks(rotation = 45)
plt.xlabel('2024')
plt.ylabel('Price USD($)')
plt.title('Change In Price Of Bitcoin In Year 2024')
plt.show()
```

This Plot shows the change in the price of Bitcoin in Year 2024
